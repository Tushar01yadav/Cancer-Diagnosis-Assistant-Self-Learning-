```python
feature_names = [ 'radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']
```

```python
df = pd.DataFrame(data.data, columns=feature_names)
df['target'] = data.target
df.head()
```

```python
sns.countplot(x='target', data=df)
plt.xticks([0, 1], ['Malignant (0)', 'Benign (1)'])
plt.title("Class Distribution")
plt.show()
```

```python
plt.figure(figsize=(15, 12))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
correlation_with_target = df.corr()['target'].drop('target')
print(correlation_with_target)
```

```python
threshold = 0.4
important_features = correlation_with_target[correlation_with_target.abs() >= threshold].index.tolist()
print(f"Features with correlation >= {threshold} or <= -{threshold}:")
print(important_features)
len(important_features)
```

# feature selection

```python
df_important = df[important_features]
corr_matrix = df_important.corr().abs()
# Upper triangle matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]
reduced_features = [f for f in important_features if f not in to_drop]

print("Dropped due to high multicollinearity:", to_drop)
print("Remaining after removing multicollinear features:", reduced_features)
len(reduced_features)
```

```python
new_df_var = reduced_features +['target']
df = df[new_df_var]
df.shape
df.head()
```
